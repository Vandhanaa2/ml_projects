```python
#separating data for analysis
legit=credit_card[credit_card.Class==0] #if 'Class' column value is 0, then that row is legit
fraudulent=credit_card[credit_card.Class==1]
print(legit.shape, fraudulent.shape)
```

Using Under-sampling method to build a sample dataset comtaing similar distribution of normal transaction and the fraudulent transaction and create a balanced dataset

No. of fraud transactions-->492

taking 492 samples randomly from legit datapoints, then concatenate the dfs

```python
legit_sample=legit.sample(n=492)
#concatenate two dataframes
new_df=pd.concat([legit_sample,fraudulent], axis=0) # combining them row-wise by axis=0
new_df.tail()
```

```python
new_df.groupby('Class').mean()
#mean values are close in both classes, thus the new_df is good
```

Splitting the data into Targets and Features(labels)

Class is the target

```python
X=new_df.drop(columns='Class', axis=1) #drop the target(Class column)
Y=new_df['Class']
print(X,Y)
```

Split the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2) #20% dataset is used for testing. The taregt of testing data i.e. Class column is the Y_test, other column datapoints are Y_train
print(X_train.shape,X_test.shape)
```

Model training

```python
model=LogisticRegression()
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy score

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data',training_data_accuracy)
```

```python
#accuracy on testing data
X_tets_prediction=model.predict(X_test)
testing_data_Accuracy=accuracy_score(X_tets_prediction,Y_test)
print('Accuracy on training data',testing_data_Accuracy)
#both training and tetsing data gives similar accuracies, that is no over or under-fitting
```
